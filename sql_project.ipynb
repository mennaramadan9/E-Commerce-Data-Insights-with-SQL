# Questions

## Customer & Order Insights

### 1. **What is the total number of orders placed by each customer?**

```python
query = '''
    SELECT customer_id, COUNT(order_id) total_orders
    FROM orders 
    GROUP BY customer_id;
'''
pd.read_sql(query, conn)
```

### 2. **What are the top 5 states with the highest number of orders?**

```python
query = '''
    SELECT customers.customer_state, COUNT(orders.order_id) num_orders
    FROM orders
    JOIN customers
        ON orders.customer_id = customers.customer_id
    GROUP BY customers.customer_state
    ORDER BY COUNT(orders.order_id) DESC
    LIMIT 5;
'''
pd.read_sql(query, conn)
```

### 3. **How many orders have been delivered late based on the estimated delivery date?**

```python
query = '''
    SELECT COUNT(order_id) num_late_orders
    FROM orders
    WHERE order_delivered_customer_date > order_estimated_delivery_date;
'''
pd.read_sql(query, conn)
```

### 4. **What is the average order value per customer?**

```python
query = '''
    SELECT orders.customer_id, AVG(order_payments.payment_value) AOV
    FROM orders
    JOIN order_payments
        ON orders.order_id = order_payments.order_id
    GROUP BY orders.customer_id;
'''
pd.read_sql(query, conn)
```

### 5. **What is the trend of monthly orders over the past year?**

```python
query = '''
    SELECT 
        strftime('%Y-%m', order_purchase_timestamp) AS order_month, 
        COUNT(order_id) AS total_orders
    FROM orders
    GROUP BY order_month
    HAVING order_month LIKE "2018%"
    ORDER BY order_month;
'''
pd.read_sql(query, conn)
```

## Product & Sales Analysis

### 6. **Which product categories generate the highest revenue?**

```python
query = '''
    SELECT pt.product_category_name_english, SUM(op.payment_value) revenue
    FROM products p
    JOIN order_items o
        ON p.product_id = o.product_id
    JOIN order_payments op
        ON o.order_id = op.order_id
    JOIN product_category_name_translation pt
        ON p.product_category_name = pt.product_category_name
    GROUP BY p.product_category_name
    ORDER BY revenue DESC;
'''
pd.read_sql(query, conn)
```

### 7. **What is the average shipping time per product category?**

```python
query = '''
    SELECT 
        pt.product_category_name_english, 
        AVG(julianday(os.order_delivered_customer_date) - julianday(os.order_purchase_timestamp)) avg_shipping_time
    FROM products p
    JOIN order_items o
        ON p.product_id = o.product_id
    JOIN orders os
        ON o.order_id = os.order_id
    JOIN product_category_name_translation pt
        ON p.product_category_name = pt.product_category_name
    WHERE os.order_delivered_customer_date IS NOT NULL
    GROUP BY p.product_category_name
    ORDER BY avg_shipping_time DESC;
'''
pd.read_sql(query, conn)
```

### 8. **Which sellers have the highest number of sales?**

```python
query = '''
    SELECT seller_id, COUNT(order_id) sales
    FROM order_items
    GROUP BY seller_id
    ORDER BY sales DESC
    LIMIT 10;
'''
pd.read_sql(query, conn)
```

### 9. **What is the average product price in each category?**

```python
query = '''
    SELECT 
        pt.product_category_name_english,
        AVG(oi.price) average_price
    FROM products p
    JOIN order_items oi
        ON p.product_id = oi.product_id
    JOIN product_category_name_translation pt
        ON p.product_category_name = pt.product_category_name
    WHERE oi.price > 0
    GROUP BY p.product_category_name
    ORDER BY average_price DESC;
'''
pd.read_sql(query, conn)
```

### 10. **What percentage of products have more than 3 photos?**

```python
query = '''
    SELECT 
        (CAST(COUNT(product_id) AS REAL) / (SELECT COUNT(product_id) FROM products)) * 100 perc_products
    FROM products p
    WHERE product_photos_qty > 3;
'''
pd.read_sql(query, conn)
```

## Payment & Revenue Analysis

### 11. **What is the total revenue generated from all orders?**

```python
query = '''
    SELECT SUM(payment_value) total_revenue
    FROM order_payments;
'''
pd.read_sql(query, conn)
```

### 12. **What is the most common payment type used by customers?**

```python
query = '''
    SELECT op.payment_type, COUNT(op.payment_type) payment_type_count
    FROM order_payments op
    WHERE op.payment_type != "not_defined"
    GROUP BY op.payment_type
    ORDER BY COUNT(op.payment_type) DESC;
'''
pd.read_sql(query, conn)
```

### 13. **How many orders are paid in installments, and what is the average number of installments per order?**

```python
query = '''
    SELECT 
        COUNT(order_id) installments_count, 
        AVG(op.payment_installments) avg_installments
    FROM order_payments op
    WHERE op.payment_installments > 1;
'''
pd.read_sql(query, conn)
```

### 14. **What is the distribution of payment values across different states?**

```python
query = '''
    SELECT 
        c.customer_state, 
        SUM(op.payment_value) total_payment_value
    FROM order_payments op
    JOIN orders o
        ON op.order_id = o.order_id
    JOIN customers c
        ON o.customer_id = c.customer_id
    GROUP BY c.customer_state
    ORDER BY total_payment_value DESC;
'''
pd.read_sql(query, conn)
```

### 15. **What is the total freight cost per seller?**

```python
query = '''
    SELECT 
        oi.seller_id, 
        SUM(oi.freight_value) total_freight_cost
    FROM order_items oi
    GROUP BY oi.seller_id
    ORDER BY total_freight_cost DESC;
'''
pd.read_sql(query, conn)
```

## Review & Customer Satisfaction

### 16. **What is the average review score per seller?**

```python
query = '''
    SELECT oi.seller_id, AVG(ow.review_score) avg_score
    FROM order_items oi
    JOIN order_reviews ow
        ON oi.order_id = ow.order_id
    GROUP BY oi.seller_id
    ORDER BY avg_score DESC;
'''
pd.read_sql(query, conn)
```

### 17. **Which products have received the lowest ratings?**

```python
query = '''
    SELECT oi.product_id, ow.review_score rating
    FROM order_reviews ow
    JOIN order_items oi
        ON ow.order_id = oi.order_id
    GROUP BY oi.product_id
    ORDER BY ow.review_score
    LIMIT 10;
'''
pd.read_sql(query, conn)
```

### 18. **How long does it take on average for a review to be answered?**

```python
query = '''
    SELECT 
        AVG(julianday(review_answer_timestamp) - julianday(review_creation_date)) answer_period
    FROM order_reviews
    WHERE review_answer_timestamp IS NOT NULL;
'''
pd.read_sql(query, conn)
```

### 19. **What is the percentage of orders that have received a review?**

```python
query = '''
    SELECT 
        COUNT(DISTINCT orv.order_id) * 100.0 / COUNT(DISTINCT o.order_id) perc_orders_reviewed
    FROM orders o
    LEFT JOIN order_reviews orv
        ON o.order_id = orv.order_id;
'''
pd.read_sql(query, conn)
```

## Leads & Business Development

### 21. **How many marketing-qualified leads (MQLs) have converted into closed deals?**

```python
query = '''
    SELECT COUNT(mql_id) count_leads_closed
    FROM leads_closed;
'''
pd.read_sql(query, conn)
```

### 22. **Which business segment has the highest lead conversion rate?**

```python
query = '''
    SELECT business_segment, COUNT(mql_id) 
    FROM leads_closed
    GROUP BY business_segment
    ORDER BY COUNT(mql_id) DESC;
'''
pd.read_sql(query, conn)
```

### 23. **What is the average time taken to close a lead after first contact?**

```python
query = '''
    SELECT AVG(JULIANDAY(lc.won_date) - JULIANDAY(lq.first_contact_date)) avg_close_time
    FROM leads_closed lc
    JOIN leads_qualified lq
        ON lc.mql_id = lq.mql_id;
'''
pd.read_sql(query, conn)
```

### 24. **How many sellers are from each business type?**

```python
query = '''
    SELECT business_type, COUNT(seller_id) num_of_sellers
    FROM leads_closed
    WHERE business_type IS NOT NULL
    GROUP BY business_type
    ORDER BY num_of_sellers DESC;
'''
pd.read_sql(query, conn)
```

### 25. **Which sellers have the highest declared monthly revenue?**

```python
query = '''
    SELECT seller_id, declared_monthly_revenue
    FROM leads_closed
    GROUP BY seller_id
    ORDER BY declared_monthly_revenue DESC
    LIMIT 10;
'''
pd.read_sql(query, conn)
```
